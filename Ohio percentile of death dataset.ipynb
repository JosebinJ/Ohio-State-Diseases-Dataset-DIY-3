```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({'Cause of deaths':['Chronic Disease', 'Unintetional Injuries', 'Alzheimers', 'Infuenza and Pneumonia', 'Sepsis', 'Others'],'Percentile':[62, 5, 4, 2, 1 , 26 ] })
```

```python
df
```

```python
plt.pie(df.Percentile, labels=df['Cause of deaths'], explode=[0,0.1,0.2,0.3,0.4,0],autopct='%1.1f%%')
plt.title('Percentile of deaths caused by various diseases')
plt.show()
```

```python
name=['Chronic Disease', 'Unintetional Injuries', 'Alzheimers', 'Infuenza and Pneumonia', 'Sepsis', 'Others']
```

```python
df.plot.pie(y='Percentile',labels=name,explode=[0,0.1,0.2,0.3,0.4,0],autopct='%.0f%%')
```

```python
plt.pie(df.Percentile, labels=df['Cause of deaths'], explode=[0,0.1,0.2,0.3,0.4,0],autopct='%.0f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.title('Percentile of deaths caused by various diseases')
plt.show()
```
